## 📌 Hipótesis 1 – Evolución del cuerpo del sprinter

**Objetivo:**  
Analizar cómo han cambiado la edad, altura y peso promedio de los medallistas en la prueba de 100 metros a lo largo de la historia olímpica.

```python
# 🏅 Proyecto Olímpico – Análisis de Estadísticas Descriptivas
# Filtrar medallistas 100 metros
sprint = df[(df['Event'].str.contains("100 metres")) & (df['Medal'].notna())]

# Calcular promedios por año
promedios = sprint.groupby('Year')[['Age', 'Height', 'Weight']].mean()

# Graficar evolución
fig, axs = plt.subplots(3, 1, figsize=(8, 12))
for i, col in enumerate(['Age', 'Height', 'Weight']):
    axs[i].plot(promedios.index, promedios[col], marker='o')
    axs[i].set_title(f'{col} promedio por año')
    axs[i].set_xlabel('Año')
    axs[i].set_ylabel(col)
plt.tight_layout()
plt.show()

promedios
```

Hallazgos:

Altura promedio: aumentó de ~173 cm en 1896 a más de 182 cm en 2016.

Peso promedio: creció de ~63–66 kg a mediados del siglo XX a ~74 kg en 2016 (+10 kg).

Edad promedio: se mantuvo entre 20 y 24 años, con leve aumento en las últimas décadas.

Conclusión:
La hipótesis queda confirmada. Se observa una clara evolución física en los medallistas de 100 m, sobre todo en altura y peso, lo que respalda la idea de una creciente especialización y optimización física.

## 📌 Hipótesis 3 – Cambios culturales y geográficos (100m – medallistas)

**Objetivo:** visualizar cómo varía la representación de países en el medallero de 100m a lo largo del tiempo y medir la diversidad.

**Qué haremos:**
1) Contar medallas de 100m por país y año.  
2) Quedarnos con los **8 países** más representativos y agrupar el resto en **“Otros”**.  
3) Hacer un **stacked bar chart** por año (y otra versión en **porcentaje**).  
4) Calcular **diversidad por década** (número de países distintos) y **participación máxima** del país dominante por década.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1) Filtrar medallistas de 100m y contar medallas por país y año
sprint_medals = df[
    df['Event'].str.contains("100 metres", case=False, na=False) & df['Medal'].notna()
]
year_team = sprint_medals.groupby(['Year','Team']).size().reset_index(name='Medals')

# 2) Top 8 países históricos y “Otros”
top_teams = (
    year_team.groupby('Team')['Medals']
    .sum()
    .sort_values(ascending=False)
    .head(8)
    .index
)

year_team_top = year_team.copy()
year_team_top['Team'] = np.where(year_team_top['Team'].isin(top_teams), year_team_top['Team'], 'Otros')

pivot_counts = (
    year_team_top
    .groupby(['Year','Team'])['Medals']
    .sum()
    .unstack(fill_value=0)
    .sort_index()
    .astype(int)
)

# 3A) Barras apiladas: cantidad absoluta de medallas por año
ax = pivot_counts.plot(kind='bar', stacked=True, figsize=(12,6), width=0.9)
ax.set_title('100m – Medallas por año (Top países + Otros)')
ax.set_xlabel('Año')
ax.set_ylabel('N° de medallas')
ax.legend(title='País', bbox_to_anchor=(1.02, 1), loc='upper left')
plt.tight_layout()
plt.show()

# 3B) Barras apiladas normalizadas (% por año)
pivot_pct = pivot_counts.div(pivot_counts.sum(axis=1), axis=0).fillna(0)

ax = pivot_pct.plot(kind='bar', stacked=True, figsize=(12,6), width=0.9)
ax.set_title('100m – Distribución porcentual de medallas por año (Top países + Otros)')
ax.set_xlabel('Año')
ax.set_ylabel('Participación (%)')
ax.legend(title='País', bbox_to_anchor=(1.02, 1), loc='upper left')
plt.tight_layout()
plt.show()

# 4) Métricas de diversidad por década
tmp = year_team_top.copy()
tmp['Decade'] = (tmp['Year'] // 10) * 10

# Número de países distintos con medalla por década
unique_teams_decade = tmp.groupby('Decade')['Team'].nunique().rename('UniqueCountries')

# Participación del país dominante por década (share máximo)
share_dom = (
    tmp.groupby(['Decade','Team'])['Medals'].sum()
    .groupby(level=0)
    .apply(lambda s: (s / s.sum()).max())
    .rename('DominantCountryShare')
)

summary_diversity = pd.concat([unique_teams_decade, share_dom], axis=1)
summary_diversity
```

```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# --- Datos base (medallistas 100m) ---
# Si ya tenés df, omití este read_csv
# df = pd.read_csv("athlete_events.csv")

sprint = df[
    df['Event'].str.contains("100 metres", case=False, na=False)
    & df['Medal'].notna()
][['Year','Age','Height','Weight','Team']].dropna()

# --- Correlaciones Pearson que pide el hito ---
pares = [('Year','Height'), ('Year','Weight'), ('Height','Weight'), ('Age','Height'), ('Age','Weight')]
print("Correlaciones Pearson (r, p-valor, n):")
resultados_corr = {}
for a,b in pares:
    sub = sprint[[a,b]].dropna()
    r,p = pearsonr(sub[a], sub[b])
    resultados_corr[(a,b)] = (r,p,len(sub))
    print(f"- {a} vs {b}: r={r:.3f}, p={p:.3g}, n={len(sub)}")

# --- Regresión lineal: tendencia temporal de altura y peso ---
def fit_lineal(ycol):
    sub = sprint[['Year', ycol]].dropna()
    X = sub[['Year']]
    y = sub[ycol]
    m = LinearRegression().fit(X, y)
    r2 = m.score(X, y)
    print(f"\nRegresión {ycol} ~ Year: y = {m.intercept_:.2f} + {m.coef_[0]:.3f}*Year  |  R²={r2:.3f}")
    # Predicciones a futuro (ejemplo)
    anios = pd.DataFrame({'Year':[2020, 2024, 2028]})
    preds = m.predict(anios)
    print(f"Predicciones {ycol} en 2020/2024/2028:", [round(v,2) for v in preds])

    # Gráfico con recta
    plt.figure(figsize=(6,4))
    plt.scatter(X, y, s=20, alpha=0.6)
    xs = np.linspace(X.min()[0], X.max()[0], 100)
    plt.plot(xs, m.predict(pd.DataFrame({'Year':xs})))
    plt.title(f"{ycol} vs Year (con recta de regresión)")
    plt.xlabel("Year"); plt.ylabel(ycol); plt.tight_layout(); plt.show()

fit_lineal('Height')
fit_lineal('Weight')
```

```python
import pandas as pd

# Filtrar 100m medallistas
sprint = df[
    df['Event'].str.contains("100 metres", case=False, na=False)
    & df['Medal'].notna()
]

# ==========================
# 1) IDM - Índice de Diversidad del Medallero
# ==========================
idms = []
for year, group in sprint.groupby('Year'):
    paises_con_medalla = group['Team'].nunique()
    paises_totales = df[(df['Year'] == year) & (df['Event'].str.contains("100 metres", case=False, na=False))]['Team'].nunique()
    idm = paises_con_medalla / paises_totales if paises_totales > 0 else None
    idms.append({'Year': year, 'IDM': idm})

idm_df = pd.DataFrame(idms)
print("IDM por año:\n", idm_df)

# ==========================
# 2) TCA - Tasa de Cambio Antropométrico
# ==========================
sprint_stats = sprint.groupby('Year')[['Height','Weight']].mean().reset_index()

# Altura
tca_altura = (sprint_stats['Height'].iloc[-1] - sprint_stats['Height'].iloc[0]) / (sprint_stats['Year'].iloc[-1] - sprint_stats['Year'].iloc[0])

# Peso
tca_peso = (sprint_stats['Weight'].iloc[-1] - sprint_stats['Weight'].iloc[0]) / (sprint_stats['Year'].iloc[-1] - sprint_stats['Year'].iloc[0])

print(f"\nTCA Altura (cm/año): {tca_altura:.3f}")
print(f"TCA Peso (kg/año): {tca_peso:.3f}")
```
